```python
import librosa
import pandas as pd
import numpy as np
import librosa.display
import matplotlib.pyplot as plt
import os
```

```python
def save_mel_spec(audio,image):
    audio_data ,sample_rate = librosa.load(audio)
    mel_spec = librosa.feature.melspectrogram(y=audio_data,sr = sample_rate,n_fft=1024)
    log_mel_spec = librosa.power_to_db(mel_spec)
    plt.figure(figsize=(14,5))
    librosa.display.specshow(log_mel_spec,sr=sample_rate)
    plt.savefig(image)
    plt.close()
```

```python
audio_dir = '../Dataset'
image_dir = '../Image'
for audio in os.listdir(os.path.join(audio_dir,'AD')):
    audio_path = os.path.join(audio_dir,'AD',audio)
    image = audio.replace('.wav','.png')
    image_path = os.path.join(image_dir,'Dementia',image)
    save_mel_spec(audio_path,image_path)
```

```python
for audio in os.listdir(os.path.join(audio_dir,'Control')):
    audio_path = os.path.join(audio_dir,'Control',audio)
    image = audio.replace('.wav','.png')
    image_path = os.path.join(image_dir,'Control',image)
    save_mel_spec(audio_path,image_path)
```
