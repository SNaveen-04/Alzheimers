```python
import tensorflow as tf
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.applications.resnet50 import preprocess_input
from tensorflow.keras.preprocessing import image
import numpy as np
```

```python
model = ResNet50(weights=None,include_top=False)
```

```python
def extract_feature(model,image_path):
    img = image.load_img(image_path,target_size=(228,228))
    img = image.img_to_array(img)
    img = np.expand_dims(img,axis=0)
    img = preprocess_input(img)
    feature = model.predict(img)
    return feature
```

```python
import os
data_dir = '../TrainSet'
data_label = os.listdir(data_dir)
```

```python
data_label
```

```python
import pandas as pd
features = []
labels = []
for label_name in data_label:
    if (label_name=='Control'):
        label = 0
    else:
        label = 1
    for data in os.listdir(os.path.join(data_dir,label_name)):
        image_path = os.path.join(data_dir,label_name,data)
        feature = extract_feature(model,image_path)
        features.append(feature)
        labels.append(label)

data = {'feature':features,'label':labels}
datadf = pd.DataFrame(data)
datadf.to_csv('../CSV/resnet.csv')
```
